```python
SalesPromotionID = []
SalesPromotionName = []
startdate = []
enddate = []
SalesPromotion_description = []
```

```python
def num_to_char(num):
    """数字转中文"""
    num=str(num)
    new_str=""
    num_dict={"0":u"零","1":u"一","2":u"二","3":u"三","4":u"四","5":u"五","6":u"六","7":u"七","8":u"八","9":u"九"}
    listnum=list(num)
    # print(listnum)
    shu=[]
    for i in listnum:
        # print(num_dict[i])
        shu.append(num_dict[i])
    new_str="".join(shu)
    # print(new_str)
    return new_str
```

```python
import datetime
```

```python
for i in range(nums):
    date_1 = f.date_between()
    startdate.append(date_1) 
    delta=datetime.timedelta(days=np.random.randint(1,31))
    date_2 = date_1 + delta
    enddate.append(date_2)  
    discount_1 = np.random.randint(1,10)
    discount = discount_1 /10 #random.number(0, 1, 0.01)#randomFloat(2, 0, 1)# 生成浮点数，两位小数点，范围是0-1之间
    SalesPromotion_description.append(discount)
    SalesPromotionID.append(f.iban())
    name = num_to_char(discount_1)#变成文字
    SalesPromotionName.append(name + '折')
```

```python
SalesPromotion['SalesPromotionID'] = SalesPromotionID
SalesPromotion['name'] = SalesPromotionName
SalesPromotion['startdate'] = startdate
SalesPromotion['enddate'] = enddate
SalesPromotion['name'] = SalesPromotionName
SalesPromotion['description'] = SalesPromotion_description
```
